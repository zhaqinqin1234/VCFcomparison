### Common variants

### Identify different rows between files

```python
# Combine both dataframes with an indicator
df_all = pd.concat([df3.assign(Source='file1'), df4.assign(Source='file2')], ignore_index=True)

# Identify unique rows and keep the source information
unique_rows = df_all.drop_duplicates(subset=["GeneSymbol",'Cdot','PdotOneLetter','FileName-Re'], keep=False)

# Extract rows unique to each file
unique_file1 = unique_rows[unique_rows['Source'] == 'file1']
unique_file2 = unique_rows[unique_rows['Source'] == 'file2']
```

```python
# Save unique rows in file1
unique_file1.to_csv('unique_previousVariant.csv', index=False)

# Save unique rows in file2
unique_file2.to_csv('unique_newVariants.csv', index=False)
```
